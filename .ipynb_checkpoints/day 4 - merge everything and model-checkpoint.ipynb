```python
df_train.drop('Unnamed: 0', axis=1, inplace=True)
df_test.drop(['Unnamed: 0', 'Id'], axis=1, inplace=True)
```

```python
df_train.set_index('Date', inplace=True)
df_train.index = pd.to_datetime(df_train.index)
df_train.sort_index(inplace=True)
df_train.head(1)
```

```python
df_test.set_index('Date', inplace=True)
df_test.index = pd.to_datetime(df_test.index)
df_test.sort_index(inplace=True)
df_test.head(1)
```

```python
df_train.drop('WnvPresent', axis=1, inplace=True)
```

```python
# drop redundant lines

cols = ['Address', 'Block', 'Street', 'AddressNumberAndStreet', 'Latitude', 'Longitude', 'AddressAccuracy']
df_train.drop(cols, axis=1, inplace=True)
df_test.drop(cols, axis=1, inplace=True)
```

```python
train_dummy = pd.get_dummies(df_train['Species'], drop_first=True)
test_dummy = pd.get_dummies(df_test['Species'], drop_first=True)
```

```python
df_train2 = pd.concat([df_train, train_dummy], axis=1)
df_test2 = pd.concat([df_test, test_dummy], axis=1)
```

```python
df_train2.drop('Species', axis=1, inplace=True)
df_test2.drop('Species', axis=1, inplace=True)
```

```python
df_train2.head(1)
```

```python
# merge spray data

df_spray.set_index('Date', inplace=True)
df_spray.index = pd.to_datetime(df_spray.index)
df_spray.sort_index(inplace=True)
df_spray.head(1)
```

```python
date_spray=[x.replace(hour=0,minute=0,second=0,microsecond=0) for x in df_spray.index.tolist()]


def merge_spray(row):
    column_name = row[0]
    date_column = row.name
    date_column=date_column.replace(hour=0,minute=0,second=0,microsecond=0)
    if date_column in date_spray:
        index_row=date_spray.index(date_column)
        index_col=df_spray.columns.tolist().index(column_name)
        result = df_spray.iloc[index_row,index_col]
    else:
        result = 1000
    
    return result
```

```python
df_train2['spray'] = df_train2.apply(merge_spray, axis=1)
```

```python
test_trap_unique=list(df_test2.Trap.unique())
train_trap_unique=list(df_train2.Trap.unique())
not_in_train=[]

for item in test_trap_unique:
    if item not in train_trap_unique:
        not_in_train.append(item)

df_test2.loc[df_test2.Trap.isin(not_in_train),"Trap"]=df_test2.loc[df_test2.Trap.isin(not_in_train),"Trap"].apply(lambda x: x[:4])
df_test2.loc[df_test2.Trap=="T234","Trap"]="T235"
```

```python
df_test2['spray'] = df_test2.apply(merge_spray, axis=1)
```
